Create a DataFrame

```python
import pandas as pd

data = {
    "Name": ["Mansur", "Daniya", "Arailym", "Ramazan", "Artyom"],
    "Age": [19, 21, 20, 17, 18],
    "Score": [85, 92, 78, 100, 95]
}

df = pd.DataFrame(data)

df
```

Data Exploration

Display first 3 rows

```python
df.head(3)
```

Summary statistics

```python
df.describe()
```

Average score

```python
average_score = df["Score"].mean()
print(average_score)
```

Data Filtering

Students older than 20

```python
older_than_20 = df[df["Age"] > 20]
older_than_20
```

Students with score above 80

```python
score_above_80 = df[df["Score"] > 80]
score_above_80
```

```python
df.to_csv("students.csv", index=False)
```
