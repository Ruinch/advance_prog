```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students.csv")
df
```

Line Plot

```python
names = df["Name"]
scores = df["Score"]
```

Create a line plot

```python
plt.figure()
plt.plot(names, scores, marker='o')
```

Labels and title

```python
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Line Plot")
plt.show()
```
