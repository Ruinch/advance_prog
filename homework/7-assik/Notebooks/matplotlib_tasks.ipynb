```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students.csv")
df
```

Line Plot

```python
names = df["Name"]
scores = df["Score"]
```

```python
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Line Plot")

plt.plot(names, scores, marker='o')
plt.figure()
```

Bar Chart

```python
plt.bar(names, scores)
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Bar Chart")

plt.show()
plt.figure()
```
