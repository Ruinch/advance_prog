Task 2

Find sum, mean, maximum

```python
array_sum = array_2d.sum()
array_mean = array_2d.mean()
array_max = array_2d.max()

print(array_sum)
print(array_mean)
print(array_max)
```

```python
array_multiplied = array_2d * 2
print(array_multiplied)
```
